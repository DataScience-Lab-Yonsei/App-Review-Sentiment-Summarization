#4. Token

## 1. Tokenize

```python
okt = Okt()
```



*   Noun , Verb , Adjective , Determiner , Adverb 위주 사용
*   stem = False, norm = False로 설정하여 최대한 말투를 고려하도록 노력



```python
def CustomTokenizer(corpus) :
    
    tokenized = []

    stop = ['Josa','Suffix', 'Conjunction' , 'Exclamation' , 'PreEomi' , 'Eomi' , 'Punctuation', 'Foreign', 'Number' , 'KoreanParticle']

    for i, j in okt.pos(corpus, stem = False, norm = False) :

        if j in stop :

            continue

        tokenized.append(i)

    return tokenized  
```

```python
review_copy = review_df.copy()

for i in tqdm(range(len(review_copy))) : 

  review_copy.loc[i , 'review'] = CustomTokenizer(review_copy.loc[i , 'review'])

review_copy.head()
```

## 2. 토큰 개수 설정



*   토큰 개수를 10개 이상으로 설정하여 무성의한 리뷰 제거



```python
idx_list = []

for i in tqdm(range(len(review_copy))) : 

  if len(review_copy.loc[i , 'review']) >= 10 :

    idx_list.append(i)

review_copy = review_copy.loc[idx_list , :].reset_index(drop = True)

review_copy
```

** 유사 리뷰 구별 방법 **

1.   (Stem , Norm) = (F , F)으로 토큰화
2.   10개 이하의 토큰 개수를 지닌 리뷰는 무성의한 리뷰로 판단 및 제거
3.   임의의 tokenized review에 대하여 나머지 tokenized review와 자카드 유사도 계산 
3.   임의의 tokenized review에 대하여 자카드 유사도가 0.6 이상인 리뷰를 list로 저장
4.   자카드 유사도가 0.6 이상인 리뷰의 존재 여부에 따라 0,1의 값을 갖는 Jaccard_YN (binary) 변수 생성
5.   App별(BetterSleep, Calm, ... ) 전체 리뷰 중 Jaccard_YN == 1인 리뷰의 비율 확인
6.   App별(BetterSleep, Calm, ... ) 5점 리뷰 중 Jaccard_YN == 1인 리뷰의 비율 확인

# 5. 전체 리뷰

## App별 유사 리뷰 비율

```python
def jaccard_similarity(sentence_1, sentence_2) :

  s1 = set(sentence_1)
  s2 = set(sentence_2)

  if len(s1.union(s2)) != 0 :

    result = float(len(s1.intersection(s2)) / len(s1.union(s2)))

  else :

    result = 0

  return result
```

```python
review_copy_all = review_copy.copy() 

review_copy_all
```

```python
jaccard_idx = []

for i in tqdm(range(len(review_copy_all))) :

  idx_list = []

  for j in range(len(review_copy_all)) :

    if i != j and jaccard_similarity(review_copy_all.loc[i,'review'] , review_copy_all.loc[j,'review']) > 0.6 :

      idx_list.append(j)

  jaccard_idx.append(idx_list)
```

```python
review_copy_all.insert(6 , 'Jaccard' , jaccard_idx)

review_copy_all.head()
```

```python
review_copy_all['Jaccard_YN'] = list(np.repeat(1 , len(review_copy_all)))

for i in tqdm(range(len(review_copy_all))) :

  if len(review_copy_all.loc[i , 'Jaccard']) == 0 :

    review_copy_all.loc[i , 'Jaccard_YN'] = 0 

review_copy_all.head()
```

```python
review_copy_all[review_copy_all['Jaccard_YN'] == 1]
```

```python
App_prob = []

for app in list(review_copy_all['App'].unique()) :

  con_1 = review_copy_all['App'] == app

  con_2 = review_copy_all['Jaccard_YN'] == 1

  if len(review_copy_all.loc[con_1 , :]) != 0 : 
    
    prob = round(len(review_copy_all.loc[con_1 & con_2 , : ]) / len(review_copy_all.loc[con_1 , :]), 4)

  else :

    prob = 0

  prob = str(prob*100) + '%'

  App_prob.append(prob)

App_prob
```

```python
App_prob[17] = '0.94%'
```

```python
pd.DataFrame({'App' : list(review_copy_all['App'].unique()) , 'Prob' : App_prob})
```
